```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

Введем случайную величину и сгенерируем выборку

```python
rv = sts.cosine()
```

Сгенерируем выборку объема 1000

```python
sample = rv.rvs(size=1000)
```

Выведем гистограмму и теоретическую плотность распределения

```python
x_array = np.linspace(-4, 4, 100)
pdf = rv.pdf(x_array)
```

```python
plt.hist(sample, normed = True)
plt.plot(x_array, pdf)
```

Теоретическая дисперсия исходного распределения

```python
DX = 1.2899
```

Распределение выборочных средних описывается нормальным распределением с нулевым матожиданием (равным матожиданию исходного распределения) и дисперсией равной $\frac{DX}{n}$, где $n$ $-$ это размер выборок

Посторим гистограммы и теоретические кривые распределения средних значений по выборкам разного объема

```python
x_array = np.linspace(-2, 2, 100)
```

```python
sample_size = [5, 10, 50, 100]

for n in sample_size:
    sample_average = list()
    for i in range(1000):
        sample_average.append(np.mean(rv.rvs(size = n)))
    plt.figure()
    plt.hist(sample_average, normed = True)
    norm_rv = sts.norm(loc = 0, scale = math.sqrt(DX/n))
    norm_pdf = norm_rv.pdf(x_array)
    plt.plot(x_array, norm_pdf)
    plt.title('n = %d' %n)
```

Как мы можем видеть, при увеличении объема выборки точность апроксимации нормальным распределением увеличивается, а дисперсия самой выборки падает.
