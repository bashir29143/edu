```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

A = np.array([[1, 216], [1, 512], [1, 1000], [1, 1728], [1, 2197]])

b = np.array([[423], [805], [1378], [2357], [2893]])

x = la.solve(np.dot(A.T, A), np.dot(A.T, b))
x
```

```python
x_th = np.arange(6, 15, 0.01)
y_th = x[1]*x_th*x_th*x_th + x[0]
plt.plot(x_th, y_th)

x_pr = np.array([6, 8, 10, 12, 13])
y_pr = np.array([423, 805, 1378, 2357, 2893])
plt.plot(x_pr, y_pr, 'o')

plt.show()
```
